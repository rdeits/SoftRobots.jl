```julia
using PyCall
```

```julia
@pyimport scipy.spatial as spatial
```

```julia
points = [1. 0 0; 0 1 0; -1 0 0; 0 -1 0; 0 0 1; 0 0 -1]'
```

```julia
hull = spatial.ConvexHull(points')
```

```julia
hull[:simplices]
```

```julia
using PyPlot
```

```julia
cla()
hold(true)
simplices = hull[:simplices]
for i = 1:size(simplices, 1)
    plot3D([points[1,simplices[i,k]+1] for k in 1:3],
           [points[2,simplices[i,k]+1] for k in 1:3],
    zs = [points[3,simplices[i,k]+1] for k in 1:3])
end
```
